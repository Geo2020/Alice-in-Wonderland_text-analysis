```python
import os
import re
from collections import defaultdict
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
import matplotlib.pyplot as plt
from PIL import Image
```

This a prototype of a Natural Language Processing (NLP) project. Its purpose is to make a simple text analysis 
of the child book 'Alice's Adventures in Wonderland' by Lewis Carroll.

```python
def extract_book_text(text_as_str):
    """This function is specific to this text only. It cuts the preface and the end notes not written
    by the author Lewis Carroll."""
    # Split the text into lines:
    text_lines = text_as_str.split("\n")
    # Filter the lines based on the desired sentences (hardcoded for this text file):
    start_sentence = "ALICEâ€™S ADVENTURES"
    end_sentence = "              THE END"
    # Cut the text not written by Lewis Carroll:
    start_index = text_lines.index(start_sentence)
    end_index = text_lines.index(end_sentence)
    filtered_lines = text_lines[start_index:end_index]
    # Join the remaining lines back into a string:
    alice_book_text = "\n".join(filtered_lines)
    return alice_book_text
```

```python
def fetch_paragraphs_with_searched_phrase(searched_phrase):
    """For a given list of phrases (one or more), this function creates a dictionary with paragraphs, where the phrase
     is written (values) and their indexes (as keys) in the text paragraphs fetched previously (as a list of paragraphs
     """
    result_paragraphs = dict()
    for phrase in searched_phrase:
        paragraphs_dict = find_character_citations(phrase, all_paragraphs)
        for key, value in paragraphs_dict.items():
            if key not in result_paragraphs:
                result_paragraphs[key] = value
    return result_paragraphs
```

```python
def find_character_citations(character_name, text_paragraphs):
    """This function finds the occurrences (citations) of a given character in the book's paragraphs"""
    # What paragraph the input character is cited in?
    result_dict = defaultdict(list)
    character_paragraphs = dict()
    for idx, paragraph in enumerate(text_paragraphs):
        words_in_paragraph = [word for word in re.split("\W+", paragraph) if word != ""]
        if character_name in words_in_paragraph:
            # # OPTIONAL: exclude the chapter headings:
            # if "CHAPTER" in words_in_paragraph:
            #     continue

            result_dict[character_name].append(paragraph)
            character_paragraphs[idx] = paragraph
    return character_paragraphs
```

```python
def activate_sentiment_analyzer(paragraphs_dict):
    """This function analyzes the sentiments in each paragraph of an input text using the built-in module
    SentimentIntensityAnalyzer() """
    # Create an instance of the sentiment analyzer:
    sid = SentimentIntensityAnalyzer()
    # Perform sentiment analysis on a paragraph:
    sentiment_dict = {"pos": 0, "neg": 0, "neu": 0, "compound": 0}
    for paragraph in paragraphs_dict.values():
        sentiment_scores = sid.polarity_scores(paragraph)
        for key in sentiment_dict.keys():
            sentiment_dict[key] += sentiment_scores[key]
    return sentiment_dict
```

```python
def calculate_common_sentiment(sentiment_dict, paragraphs_dict):
    """This function finds the average sentiment grades for the extracted text"""
    for key in sentiment_dict.keys():
        sentiment_dict[key] /= len(paragraphs_dict)
    return sentiment_dict
```

```python
def print_common_sentiment(sentiment_dict):
    for k, v in sentiment_dict.items():
        if k == "pos":
            print(f" - positive: {round(sentiment_dict[k] * 100)}%")
        elif k == "neg":
            print(f" - negative: {round(sentiment_dict[k] * 100)}%")
        elif k == "neu":
            print(f" - neutral: {round(sentiment_dict[k] * 100)}%")
        elif k == "compound":
            if sentiment_dict[k] > 0.03:  # supposing 3% inaccuracy
                print(f" - overall positive: {round(sentiment_dict[k] * 100)}%")
            elif sentiment_dict[k] < -0.03:
                print(f" - overall negative: {round(sentiment_dict[k] * 100)}%")
            else:
                print(f" - more or less neutral: {round(sentiment_dict[k] * 100)}%")
```

```python
def remove_Dinah_from_cat_paragraphs(dictionary):
    """This function removes key-value pairs from the input dictionary under a particular condition
    in the case of 'Cheshire Cat' """
    keys_to_remove = []
    for key, value in dictionary.items():
        if "Dinah" in value:
            keys_to_remove.append(key)

    for key in keys_to_remove:
        del dictionary[key]

    return dictionary
```

```python
def print_on_console(string, symbols_per_line):
    for i in range(0, len(string), symbols_per_line):
        print(string[i:i+symbols_per_line])
```

```python
def illustrate_the_book():
    """Show some book illustrations"""
    subfolder_path = './data'
    image_files = [os.path.join(subfolder_path, file) for file in os.listdir(subfolder_path) if file.endswith('.jpg')]
    # Set the number of rows and columns for the table
    num_rows = 2
    num_cols = [3, 3]  # each row has 3 columns
    # Create a figure and axes for the table
    fig, axes = plt.subplots(num_rows, max(num_cols), figsize=(12, 8))
    for i, ax in enumerate(axes.flat):
        # Check if there are more images to display
        if i < len(image_files):
            image = Image.open(image_files[i])
            ax.imshow(image)
            ax.axis('off')
        else:
            fig.delaxes(ax)  # if there are no more images, remove the remaining empty axes
    plt.tight_layout()  # adjust the spacing between subplots
    plt.show()
```

### Read the input text file

```python
with open("data/Alice_in_Wonderland.txt", "r", encoding="utf-8") as f:
    inp_text_as_str = f.read()
```

### Preprocessing the text

```python
# Remove the text not belonging to the original book:
alice_text = extract_book_text(inp_text_as_str)
```

```python
# Remove punctuation from the text:
translator = str.maketrans('', '', string.punctuation)
cleaned_text = alice_text.translate(translator)
```

We can break down the text into smaller units, e.g. in words (tokenization).

```python
# Tokenize the text into words:
all_words = word_tokenize(cleaned_text)
print(f"\nAll cleaned words in the text without punctuation are: {len(all_words)}")
```

```python
# Define the stop words in English (articles, prepositions, conjunctions, etc.):
english_stopwords = set(stopwords.words("english"))
filtered_words = [token for token in all_words if token not in english_stopwords]  # 12,677
print(f"All filtered words (without stop words) in the text are: {len(filtered_words)}")
```

```python
# Extract all paragraphs in the book:
all_paragraphs = list(filter(None, re.split(r"[\n\n]\s*", alice_text)))
print(f"All paragraphs in the text are: {len(all_paragraphs)}")
```

Note: The character 'Cheshire cat' is named in different ways in the book - as 'Cheshire cat' or 'Cheshire Cat' 
('Cat' with first letter capitalized), or even as 'Cheshire Puss' but somewhere he is named just 'Cat' or 'cat'. 
On the other hand, there is another character which is a cat - this is Alice's pet Dinah. Thus, to find all occurrences 
of the Cheshire cat and not these of Dinah, we can apply two additional searches in the text - of the string 'Cat' 
and 'cat' and exclude from the results those, where the name 'Dinah' is given at the same paragraph.

```python
# Extract all paragraphs where 'Cheshire cat' is cited:
cheshire_cat_phrases = ["Cheshire", "Cat", "cat"]
cheshire_cat_paragraphs = fetch_paragraphs_with_searched_phrase(cheshire_cat_phrases)
cheshire_cat_paragraphs = remove_Dinah_from_cat_paragraphs(cheshire_cat_paragraphs)
cheshire_cat_paragraphs = dict(sorted(cheshire_cat_paragraphs.items()))
```

We can find the paragraphs when Alice meets the Cheshire Cat for the first and last time.

```python
first_idx_cheshire = list(cheshire_cat_paragraphs.keys())[0]
for j in range(first_idx_cheshire, 0, -1):
    if "chapter" in all_paragraphs[j].lower():
        print(f"\nAlice first meets the Cheshire Cat in {all_paragraphs[j]}, in the paragraph:")
        print_on_console(cheshire_cat_paragraphs[first_idx_cheshire], 122)  # 122 symbols per line
        break
```

```python
# In which chapter does Alice see the cat for the last time?
last_idx_cheshire = list(cheshire_cat_paragraphs.keys())[-1]
for j in range(last_idx_cheshire, 0, -1):
    if "chapter" in all_paragraphs[j].lower():
        print(f"\nAlice sees the Cheshire Cat for the last time in {all_paragraphs[j]}, in the paragraph:")
        print_on_console(cheshire_cat_paragraphs[last_idx_cheshire], 115)
        break
```

### Sentiment Analysis
Applying sentiment analysis, we can find Alice's emotions for the Cheshire cat.

```python
sentiment_cheshire = activate_sentiment_analyzer(cheshire_cat_paragraphs)
# Find the average sentiment grades for all the paragraphs:
common_sentiment_cheshire = calculate_common_sentiment(sentiment_cheshire, cheshire_cat_paragraphs)
print(f"\nAlice's sentiments toward the character 'Cheshire cat' are:")
print_common_sentiment(common_sentiment_cheshire)
```

Note: Similar information (as the above) can be obtained for another character, e.g. Queen.

```python
queen_paragraphs = find_character_citations("Queen", all_paragraphs)
queen_paragraphs = dict(sorted(queen_paragraphs.items()))

sentiment_queen = activate_sentiment_analyzer(queen_paragraphs)
common_sentiment_queen = calculate_common_sentiment(sentiment_queen, queen_paragraphs)
print(f"\nAlice's sentiments toward the character 'Queen' are:")
print_common_sentiment(common_sentiment_queen)
```

### Visualization
Also, we can visualize the occurrences (citations) of the book characters.

```python
duchess_pattern = r"Duchess"
duchess_matches = re.findall(duchess_pattern, alice_text)

rabbit_pattern = r"Rabbit"  # at the beginning, the character is with small first letter, then - with a capital
rabbit_matches = re.findall(rabbit_pattern, alice_text, re.IGNORECASE)

march_hare_pattern = r"March Hare"
march_hare_matches = re.findall(march_hare_pattern, alice_text)

hatter_pattern = r"Hatter"
hatter_matches = re.findall(hatter_pattern, alice_text)

queen_pattern = r"Queen"
queen_matches = re.findall(queen_pattern, alice_text)
```

Note: 'queen_matches' above is a list of citations of the word 'Queen', which has a longer length (words count) 
than the length of the paragraphs where Queen is cited ('queen_paragraphs').

```python
king_pattern = r"King"
king_matches = re.findall(king_pattern, alice_text)
```

Note: As explained above, the character Cheshire cat appears in the text as 'Cheshire Cat', 'Cheshire cat', 
or 'Cheshire Puss' (name 'Cheshire' is never written without 'cat' or 'puss'). So, the occurrences of this character 
are related to the citations of the words 'cat' or 'puss' (with capitalized first letter or not).

```python
cheshire_cat_pattern = r"cat|puss"
cheshire_cat_text = "\n".join(list(cheshire_cat_paragraphs.values()))
cheshire_matches = re.findall(cheshire_cat_pattern, cheshire_cat_text, re.IGNORECASE)
```

```python
# Create a bar chart showing the frequency of occurrences (citations):
citations = [len(duchess_matches), len(rabbit_matches), len(cheshire_matches), len(march_hare_matches),
             len(hatter_matches), len(queen_matches), len(king_matches)]
labels = ["Duchess", "Rabbit", "Cheshire cat", "March Hare", "Hatter", "Queen", "King"]

fig_width = 9  # in inches (found by trial and error)
fig_height = 5
plt.figure(figsize=(fig_width, fig_height))
plt.bar(labels, citations)
plt.title("Citations of the characters in the book")
plt.xlabel("Phrase")
plt.ylabel("Number of Occurrences")
plt.show()
```
